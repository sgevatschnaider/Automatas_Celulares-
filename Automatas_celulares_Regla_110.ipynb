```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import display, HTML

# Definir la Regla 110
def rule_110(left, center, right):
    if left == 1 and center == 1 and right == 1:
        return 0
    elif left == 1 and center == 1 and right == 0:
        return 1
    elif left == 1 and center == 0 and right == 1:
        return 1
    elif left == 1 and center == 0 and right == 0:
        return 0
    elif left == 0 and center == 1 and right == 1:
        return 1
    elif left == 0 and center == 1 and right == 0:
        return 1
    elif left == 0 and center == 0 and right == 1:
        return 1
    elif left == 0 and center == 0 and right == 0:
        return 0

# Generar la siguiente generación
def next_generation(current_gen):
    next_gen = np.zeros_like(current_gen)
    for i in range(1, len(current_gen)-1):
        next_gen[i] = rule_110(current_gen[i-1], current_gen[i], current_gen[i+1])
    return next_gen

# Generar la fila inicial
def generate_initial_state(size, option="center"):
    """
    Genera una fila inicial.

    option:
    - "center": Una celda viva en el centro.
    - "random": Distribución aleatoria de celdas vivas y muertas.
    """
    initial_state = np.zeros(size, dtype=int)
    if option == "center":
        initial_state[size // 2] = 1
    elif option == "random":
        initial_state = np.random.randint(0, 2, size)
    return initial_state

# Configurar la animación
def animate_automaton(rule_func, size=101, generations=50, initial_option="center", save_gif=False, gif_name="automaton.gif"):
    grid = np.zeros((generations, size), dtype=int)
    grid[0] = generate_initial_state(size, option=initial_option)

    for gen in range(1, generations):
        grid[gen] = next_generation(grid[gen-1])

    fig, ax = plt.subplots()
    ax.set_axis_off()
    mat = ax.matshow(grid, cmap="binary")

    def update(frame):
        mat.set_data(grid[:frame])
        return [mat]

    ani = animation.FuncAnimation(fig, update, frames=generations, interval=200, blit=True)

    # Guardar como GIF si se especifica
    if save_gif:
        ani.save(gif_name, writer="pillow")
        print(f"Animación guardada como {gif_name}")

    plt.show()

    # Mostrar enlace de descarga si se guardó el gif
    if save_gif:
        display(HTML(f'<a href="{gif_name}" download>Haz clic aquí para descargar el GIF</a>'))

# Ejecutar la animación con opción de guardar el GIF
animate_automaton(rule_110, size=101, generations=100, initial_option="random", save_gif=True, gif_name="automaton.gif")

```
